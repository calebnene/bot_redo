```python
#top fractals

df_ma['frac_top_bool'] = np.where(
    df_ma['mid_h'] == df_ma['mid_h'].rolling(5, center=True).max(), True, False
)

df_ma['frac_top'] = np.where(
    df_ma['mid_h'] == df_ma['mid_h'].rolling(5, center=True).max(), df_ma['mid_h'], None
)

#filling the subsequent rows with the previous fractal high till the next fractal high 
df_ma['frac_top'] = df_ma['frac_top'].ffill()
```

```python
#bottom fractals 
df_ma['frac_bottom_bool'] = np.where(
    df_ma['mid_l'] == df_ma['mid_l'].rolling(5, center=True).min(), True, False
)

df_ma['frac_bottom'] = np.where(
    df_ma['mid_l'] == df_ma['mid_l'].rolling(5, center=True).min(), df_ma['mid_l'], None
)

#filling the subsequent rows with the previous fractal high till the next fractal high 
df_ma['frac_bottom'] = df_ma['frac_bottom'].ffill()
```

Identifying Trades with Fractals 

```python
def is_trade(row):
    if (row.mid_c > row.frac_top) and (row.mid_c > row.MA_200):
        return 1 #buy 
    
    if (row.mid_c < row.frac_bottom) and (row.mid_c < row.MA_200):
        return -1 #sell
    
def sl(row):
    if row.IS_TRADE == 1:
        return row['mid_c']-row['mid_l']
        

    if row.IS_TRADE == -1:
        return row['mid_h']-row['mid_c']

def tp(row):
    if row.IS_TRADE == 1:
        return row['SL'] * 1.5

    if row.IS_TRADE == -1:
        return row['SL'] * 1.5
    
#checking the gains 

def gain(row, df_ma):
    if row.IS_TRADE == 1:
        entry_price = row['mid_c']
        tp_price = entry_price + row['TP']
        sl_price = entry_price - row['SL']

        for i in range(row.name+1, len(df_ma)):
            next_row = df_ma.iloc[i]
            # Check if TP is hit
            if next_row['mid_h'] >= tp_price:
                return row['TP']  # Gain equals TP
            
            # Check if SL is hit
            if next_row['mid_l'] <= sl_price:
                return -row['SL']  # Loss equals SL
            
    
    elif row.IS_TRADE == -1:  # Sell trade
        entry_price = row['mid_c']
        tp_price = entry_price - row['TP']
        sl_price = entry_price + row['SL']
        
        # Check next rows (candles) after the current row
        for i in range(row.name + 1, len(df_ma)):
            next_row = df_ma.iloc[i]
            # Check if TP is hit
            if next_row['mid_l'] <= tp_price:
                return row['TP']  # Gain equals TP
            
            # Check if SL is hit
            if next_row['mid_h'] >= sl_price:
                return -row['SL']  # Loss equals SL

    return 0  # No gain/loss if neither TP nor SL is hit
```

```python
df_ma['IS_TRADE'] = df_ma.apply(is_trade, axis=1)
#df['SL'] = df.apply(sl, axis = 1)
#df['TP'] = df.apply(tp, axis=1)
#df['GAIN'] = df.apply(lambda row: gain(row, df), axis=1)
```
